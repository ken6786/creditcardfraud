UNDER SAMPLING

```python
#storing  real credit card details into a variable
x=credit[credit['Class']==0]
```

```python
#storing fake credit card details into a variable
y=credit[credit['Class']==1]
```

```python
#displaying number of real credit card values
x.shape
```

```python
#displaying number of fake credit card values
y.shape
```

```python
normalSamples=x.sample(n=473)
```

```python
normalSamples.shape
```

```python
newCredit=pd.concat([normalSamples,y],ignore_index=True)
```

```python
newCredit.shape
```

```python
a=newCredit.drop(["Class"],axis=1)
```

```python
a.head()
```

```python
b=newCredit["Class"]
```

```python
a_train,a_test,b_train,b_test=train_test_split(a,b,test_size=0.20,random_state=67)
```

```python
model=RandomForestClassifier(n_estimators=5)
```

```python
model.fit(a_train,b_train)
```

```python
y_pred=model.predict(a_test)
```

```python
from sklearn.metrics import accuracy_score
```

```python
accu=accuracy_score(b_test,y_pred)
```
